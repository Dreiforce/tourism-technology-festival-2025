```python
import pandas as pd

```

```python
df = pd.read_json("data.json", encoding="utf-8-sig")
```

```python
df = df[df['areas'].str.contains("Tirol")]
```

```python

# Define what "filled" means
def is_filled(x):
    if x is None:
        return False
    if isinstance(x, (str, list, dict)):
        return len(str(x).strip()) > 0
    return True

# Compute completeness score for each row
df['filled_score'] = df.applymap(is_filled).sum(axis=1)

# Sort from most to least complete
df_sorted = df.sort_values('filled_score', ascending=False)

# Show top 10 most complete rows
df_sorted[['id', 'filled_score']].head(15)
```
